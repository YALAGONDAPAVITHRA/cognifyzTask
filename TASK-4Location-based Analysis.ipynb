```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = pd.read_csv("Dataset.csv", low_memory=False)
df.head()
```

```python
# Fill missing values
df.fillna(method='ffill', inplace=True)
```

```python
# Ensure latitude & longitude are numeric
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
```

```python
# Drop rows with missing coordinates
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)
```

```python
# Scatter plot of restaurant locations
plt.figure(figsize=(10, 6))
plt.scatter(df['Longitude'], df['Latitude'], alpha=0.5, color='blue')
plt.title("Geographical Distribution of Restaurants\n",fontsize=18,color='red')
plt.xlabel("Longitude",fontsize=13,color='green')
plt.ylabel("Latitude",fontsize=13,color='green')
plt.show()
```

```python
# Group by city and analyze restaurant count
city_group = df.groupby('City').size().reset_index(name='Restaurant Count')
print("\nTop 10 Cities with Most Restaurants:\n", city_group.sort_values(by='Restaurant Count', ascending=False).head(10))
```

```python
# Calculate average rating by city
if 'Aggregate Rating' in df.columns:
    city_avg_rating = df.groupby('City')['Aggregate Rating'].mean().reset_index()
    print("\nAverage Ratings by City:\n", city_avg_rating.head(10))
else:
    print("Column 'Aggregate Rating' not found in the dataset.")
```
